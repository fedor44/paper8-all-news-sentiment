```python
from openai import OpenAI

client = OpenAI(
    api_key="sk-"
)

def sentiment_analysis(transcription):
    completion = client.chat.completions.create(
        model="gpt-4-0125-preview",
            temperature=0,
            max_tokens=1,
            messages=[
                {
                    "role": "system",
                    #"content": "As an AI with expertise in language and emotion analysis, your task is to analyze the sentiment of the following text. Please consider the overall tone of the discussion, the emotion conveyed by the language used, and the context in which words and phrases are used. Indicate whether the sentiment is generally positive, negative, or neutral, and provide brief explanations for your analysis where possible."
                    #"content": "As an AI with expertise in language and emotion analysis, your task is to analyze the sentiment of the following text. Consider the financial implications and context when determining sentiment.Indicate whether the sentiment is just positive = 0, negative = 1, or neutral = 2, and don't need to provide brief explanations for your analysis"
                    # 은행 테스트용
                    #"content": "As an AI with expertise in language and emotion analysis, your task is to analyze the sentiment of the following text. Consider the financial implications and context when determining sentiment. return a numerical value where positive is represented by 2, neutral by 1, and negative by 0. and don't need to provide brief explanations for your analysis"
                    # 위 content 수정
                    "content": "your task is to analyze the sentiment of the following financial text. return a numerical value where positive is represented by 2, neutral by 1, and negative by 0. and When analyzing the sentiment of a statement, consider the statement neutral unless there are clear emotional expressions or connotations that indicate positivity or negativity. and don't need to provide brief explanations for your analysis"
                    # financial_phrasebank
                    #"content": "your task is to analyze the sentiment of the following financial text. return a numerical value where positive is represented by 0, neutral by 2, and negative by 1. and When analyzing the sentiment of a statement, consider the statement neutral unless there are clear emotional expressions or connotations that indicate positivity or negativity. and don't need to provide brief explanations for your analysis"
                },
                {
                   "role": "user",
                    "content": transcription
               }
            ]
        )
    return completion.choices[0].message.content
```

```python
from google.colab import drive
drive.mount('/content/drive/')
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/DeepNeuralNetwork/news.csv')
df
```

```python
# 전처리
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re
import nltk
nltk.download('stopwords')

df['processed_headline'] = df['headline']

# 불용어 제거
stop_words = set(stopwords.words('english'))
df['processed_headline'] = df['processed_headline'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))

# 불필요한 문자 제거 (느낌표와 물음표는 유지)
df['processed_headline'] = df['processed_headline'].apply(lambda x: re.sub(r'[^a-zA-Z!?\'\s]', '', x))

# 일부 뉴스만 감성 분석 하기위해 추가
final_df = pd.DataFrame({'sentence':[],'label':[]})

# positive = 2, negative = 0, or neutral = 1
for i, sentence in enumerate(df['processed_headline']):
    if i < 1000:
      continue


    sentiment = sentiment_analysis(sentence)  # 여기서 'sentence'는 문자열이어야 합니다.
    final_df.loc[i-1001] = [sentence, sentiment]
    #if i == 1000:  # 0부터 시작하므로 i == 4 일 때까지가 5개입니다.
    #    break

#sentiment = sentiment_analysis('Global economic downturn leads to widespread layoffs and reduced consumer spending, impacting markets negatively')
#final_df.loc[1001] = ['Global economic downturn leads to widespread layoffs and reduced consumer spending, impacting markets negatively', sentiment]
#sentiment = sentiment_analysis('The company released its quarterly earnings report, showing a mixture of gains and losses across different sectors')
#final_df.loc[1002] = ['The company released its quarterly earnings report, showing a mixture of gains and losses across different sectors', sentiment]

final_df
```
